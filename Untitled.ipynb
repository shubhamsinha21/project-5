# Data Cleaning

```python
# check null values
df.isnull().sum()
```

### Convert output variable into numerical terms

```python
df['emotion'].unique()
```

```python
df['emotion'].unique()
# total 6 emotions

# its an array, so
unique_emotions = df['emotion'].unique()
emotion_numbers = {} # dictionary
i = 0
for emot in unique_emotions:
    emotion_numbers[emot] = i
    i +=1

df['emotion'] = df['emotion'].map(emotion_numbers)
```

### Text Cleaning

```python
# 1. converting all texts in lowercase
df['text'] = df['text'].apply(lambda x: x.lower())
```

```python
# 2. remove punctuations
import string

def remove_punc(txt):
    return txt.translate(str.maketrans('','', string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punc)
```

```python
# 3. remove numbers
def remove_numbers(txt):
    new = ''
    for i in txt:
        if not i.isdigit():
            new = new + i
    return new
df['text'] = df['text'].apply(remove_numbers)
```

```python
# 4. remove URL'S or links -> you can create a function and remove
# 5. remove html tags -> data from kaggle, hence no html tags
```

```python
# 6. remove emojis and special values
def remove_emojis(txt):
    new = ''
    for i in txt:
        if i.isascii(): # isascii -> function/method whic has all ascii values and all emojis are created using ascii values
            new = new + i
    return new
df['text'] = df['text'].apply(remove_emojis)
```

```python
# remove stopwords
# to do this task, we have a famous lib called - nltk

import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize # stopwords to remove ke liye we will tokenize words
```

# need to download once all these stuffs
nltk.download('punkt') # -> this is used for tokenization
nltk.download('punkt_tab')
nltk.download('stopwords')

```python
stop_words = set(stopwords.words('english')) # we have all language stopwords now
len(stop_words)
```

```python
# stopwords checking
df.loc[1]['text']
```

```python
# take all text one by one and perform tokenization in all sentences
def remove(txt):
    words = word_tokenize(txt)
    cleaned = []
    for i in words:
        if not i in stop_words:
            cleaned.append(i)
    return ' '.join(cleaned)
            
```

```python
# applying function on text
df['text'] = df['text'].apply(remove)
```
